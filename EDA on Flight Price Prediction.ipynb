## Sorting date and time features

### Converting to datetime datatype

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
```

## Extracting: Year, Month and Day
 

```python
df['Year'] = df['Date_of_Journey'].dt.year
df['Month'] = df['Date_of_Journey'].dt.month
df['Day'] = df['Date_of_Journey'].dt.day
```

```python
df.info()
```

### Dropping off Day of Journey column

```python
df.drop(['Date_of_Journey'], axis=1, inplace= True)

```

```python
df.info()
```

### Converting to datetime datatype and extracting hour and minutes from Dep_Time


```python
df['Dep_Time'] = pd.to_datetime(df['Dep_Time'])
df['Dep_hour'] = df['Dep_Time'].dt.hour
df['Dep_min'] = df['Dep_Time'].dt.minute

```

### Dropping off Dep_time column

```python
df.drop(['Dep_Time'], axis=1, inplace= True)
```

```python
df.info()
```

### Converting to datetime datatype and extracting hour and minutes from Arrival_Time

```python
df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time'])
df['Arrival_hour'] = df['Arrival_Time'].dt.hour
df['Arrival_min'] = df['Arrival_Time'].dt.minute
df.drop(['Arrival_Time'], axis=1, inplace= True)


```
